### Resolucion

### 1. Power Query (30%)

#### 1.1 ¿Cuál es la diferencia entre un modo de importación vs DirectQuery?
El **modo Importación** carga los datos al modelo de Power BI, ofreciendo mejor rendimiento y mayor flexibilidad en transformaciones y DAX, pero requiere actualizaciones programadas.  
**DirectQuery** consulta los datos directamente en la fuente en tiempo real, es útil para grandes volúmenes de datos, aunque presenta limitaciones en transformaciones y depende del rendimiento del origen.

#### 1.2 ¿Cuáles son los diferentes orígenes de datos con los que has trabajado?
He trabajado con **Excel, SQL Server, SharePoint, archivos CSV y APIs Web**, integrándolos mediante procesos ETL en Power Query para análisis y reportes de BI.

#### 1.3 ¿Cuál es la diferencia entre una tabla de hechos y una de dimensiones?
La **tabla de hechos** contiene métricas numéricas (ventas, ingresos, cantidades) y claves foráneas.  
Las **tablas de dimensiones** contienen atributos descriptivos (cliente, producto, tiempo) que contextualizan los datos de la tabla de hechos.

#### 1.4 ¿Cuál es la diferencia entre un modelo estrella y copo de nieve?
El **modelo estrella** tiene una tabla de hechos central con dimensiones desnormalizadas, lo que mejora el rendimiento y la simplicidad.  
El **modelo copo de nieve** normaliza las dimensiones en varias tablas, reduce redundancia pero incrementa la complejidad de consultas.

#### 1.5 ¿En Power Query qué son las funciones y cómo las has utilizado?
Las **funciones en Power Query (lenguaje M)** encapsulan lógica reutilizable para limpieza, validación y cálculos personalizados. Las he utilizado para estandarizar transformaciones repetitivas en múltiples consultas.

#### 1.6 ¿Para qué se usa la opción de Merge y Append en Power Query?
- **Merge**: combina tablas horizontalmente usando claves comunes (similar a JOIN en SQL).
- **Append**: une tablas verticalmente apilando filas con la misma estructura (similar a UNION).

### 2. Power View (20%)

#### 2.1 ¿Qué son los bookmarks y cómo los has utilizado?
Los **bookmarks** guardan el estado de una página (filtros, selecciones, visuales) y se utilizan para crear navegación interactiva y storytelling en dashboards.

#### 2.2 ¿Qué son los field parameters y cómo los has utilizado?
Los **field parameters** permiten cambiar dinámicamente dimensiones o medidas en un visual mediante un slicer, facilitando reportes flexibles sin duplicar gráficos.

#### 2.3 ¿Qué es el formato dinámico y para qué sirve?
El **formato dinámico** permite cambiar colores, etiquetas o formatos de un visual según reglas o contexto, mejorando la legibilidad y experiencia del usuario.

#### 2.4 ¿Qué tipo de gráfico utilizarías para…?
- **Participación de ventas por dimensión**: Gráfico de donut o treemap.  
- **Correlación entre dos variables**: Gráfico de dispersión (scatter plot) con línea de tendencia.  
- **Histórico de ventas**: Gráfico de líneas o áreas con eje temporal.

#### 2.5 ¿Cuál es la diferencia entre filters y slicers?
Los **filters** aplican filtros a nivel de visual, página o reporte y suelen ser internos.  
Los **slicers** son visuales interactivos que permiten al usuario filtrar los datos directamente.

#### 2.6 ¿En qué casos utilizarías el gráfico Waterfall Chart?
Se utiliza para mostrar variaciones acumulativas positivas y negativas, como el análisis de crecimiento, márgenes o explicación de cambios entre periodos.

#### 2.7 ¿En qué casos utilizarías el gráfico Q&A?
Para exploración rápida mediante lenguaje natural, permitiendo a los usuarios generar visuales automáticamente sin conocimientos técnicos.

### 3. DAX (40%)

#### 3.1 ¿Qué contextos de evaluación influyen en DAX y en qué consiste cada uno?
- **Row Context**: evaluación fila por fila, común en funciones iteradoras.  
- **Filter Context**: definido por filtros, slicers y CALCULATE.  
- **Query Context**: generado por la consulta que solicita el resultado.

#### 3.2 ¿Es necesario tener una tabla dimensión tiempo? ¿Por qué?
Sí, porque habilita funciones de inteligencia de tiempo como YTD, MTD y comparaciones interanuales, asegurando cálculos correctos y eficientes.

#### 3.3 ¿Qué funciones de iteración conoces y cómo las aplicarías?
Funciones como **SUMX, AVERAGEX, MAXX** permiten evaluar expresiones fila por fila.  
Ejemplo:  
`SUMX(Ventas, Ventas[Precio] * Ventas[Cantidad])`

#### 3.4 ¿En qué consiste la función CALCULATE y cómo la utilizarías?
**CALCULATE** modifica el filter context para evaluar una medida bajo nuevas condiciones.  
Ejemplo:  
`CALCULATE([Ventas], Producto[Categoría] = "A")`

#### 3.5 ¿Cuál es la diferencia entre ALL, ALLSELECTED y ALLEXCEPT?
- **ALL**: elimina todos los filtros.  
- **ALLSELECTED**: respeta filtros externos del usuario.  
- **ALLEXCEPT**: mantiene filtros específicos y elimina el resto.

#### 3.6 ¿Cuál es la diferencia entre FILTER y VALUES?
- **FILTER** devuelve una tabla filtrada por una condición lógica.  
- **VALUES** devuelve valores únicos según el contexto actual.

#### 3.7 ¿En qué consiste la función KEEPFILTERS y cómo la utilizarías?
**KEEPFILTERS** mantiene los filtros existentes al usar CALCULATE, evitando que sean reemplazados por nuevos filtros.

#### 3.8 ¿Qué funciones de inteligencia de tiempo conoces y cómo las aplicarías?
Funciones como **TOTALYTD, SAMEPERIODLASTYEAR, DATESINPERIOD** permiten cálculos acumulados y comparaciones temporales.

#### 3.9 ¿Cuál es la diferencia entre USERELATIONSHIP, CROSSFILTER y TREATAS?
- **USERELATIONSHIP**: activa una relación inactiva.  
- **CROSSFILTER**: cambia la dirección de filtrado.  
- **TREATAS**: aplica una tabla virtual como filtro.

### 4. Power BI Service (10%)

#### 4.1 ¿Cuáles son los tipos de licencia que existen en Power BI Service?
Free, Pro, Premium Per User (PPU) y Premium Per Capacity.

#### 4.2 ¿Cuál es la diferencia entre ellas?
Free es para consumo personal, Pro permite colaboración y Premium ofrece capacidad dedicada y funcionalidades avanzadas.

#### 4.3 ¿Cuáles son los roles disponibles en un workspace?
Admin, Member, Contributor y Viewer, con distintos niveles de permisos.

#### 4.4 ¿Qué es RLS?
**Row-Level Security** restringe el acceso a filas de datos según el usuario.

#### 4.5 ¿Qué es un Gateway? ¿Cuántos tipos hay y cómo funcionan?
Es un puente seguro entre Power BI Service y datos on-premise.  
Tipos:
- **Personal**: uso individual.  
- **Standard**: uso corporativo con acceso compartido y refrescos programados.

#### 4.6 ¿Qué son los dataflows en Power BI Service?
Son procesos ETL en la nube basados en Power Query que permiten reutilizar y estandarizar la preparación de datos para múltiples reportes.
